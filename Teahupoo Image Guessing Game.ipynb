```python
import matplotlib.pyplot as plt
import numpy as np
laird_hi_res = "LairdHamiltonTeahupoo.jpg"
LAIRD = plt.imread(laird_hi_res)
T = [90,60,45,36,30,24,18,15,12,9,6,3,2,1]
E = [120,80,60,48,40,32,24,20,16,12,8,4,2,1]

for n in range(0,14):

    Image = str(object = "LairdHamiltonTeahupoo.jpg")
    plt.imread(Image).shape
    tile_x = int(LAIRD.shape[1]/E[n])
    tile_y = int(LAIRD.shape[0]/T[n])

    print("{0} pixels".format(1600)) #1600 pixels
    x = int(LAIRD.shape[1]/tile_x)
    y = int(LAIRD.shape[0]/tile_y)
    z = int(LAIRD.shape[2])
    name = np.zeros([y,x,z],dtype='uint8')
    for a in range(0,y):
        for b in range(0,x):
            for c in range(0,3):
                name[a,b,c] = np.average(LAIRD[a*tile_y:(a+1)*tile_y,b*tile_x:(b+1)*tile_x,c])
    red = np.zeros(np.shape(LAIRD),dtype = 'uint8')
    for a in range(0,y):
        for b in range(0,x):
            for c in range(0,3):
                red[(a*tile_y):((a+1)*tile_y),(b*tile_x):((b+1)*tile_x),c] = name[a,b,c]
                neg = np.zeros(np.shape(LAIRD),dtype = 'uint8')
    plt.imsave("#{}LairdHamiltonTeahupoo.jpg",red)
    for a in range(0,360):
        for b in range(0,480):
            for c in range(0,3):
                neg[a,b,c] = 1 - red[a,b,c]
    laird_hi_res.replace("","#{0}low res".format(n+1),1)
    plt.imsave("#{} Laird Low Res photo.jpg".format(n+1),red)
    plt.imsave("#{} Laird neg Low Res photo.jpg".format(n+1),neg)
    print("name of this file is ","#{} LAIRD Low Res photo.jpg".format(n+1))
    print("also #{} Laird neg Low Res photo.jpg".format(n+1))
```
